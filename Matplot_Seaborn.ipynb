### Matplotlib is a python library used for data visualization

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,20)
x
```

```python
y = 5*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y)
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.show()
```

```python
plt.plot(x, y, color='g')
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.show()
```

```python
plt.plot(x, y, color='g', linewidth = 3)
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.show()
```

```python
plt.plot(x, y, color='g', linewidth = 3, linestyle = ':')
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.show()
```

```python
y2 = x*3
```

```python
plt.plot(x, y, color='g', linewidth = 3)
plt.plot(x, y2, color='r', linewidth = 3)
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.show()
```

```python
plt.plot(x, y, color='g', linewidth = 3)
plt.plot(x, y2, color='r', linewidth = 3)
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.show()
```

```python
plt.plot(x, y, color='g', linewidth = 3, linestyle = ':')
plt.plot(x, y2, color='r', linewidth = 3, linestyle = '-.')
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x, y, color='g', linewidth = 3, linestyle = ':')
plt.subplot(2,1,2)
plt.plot(x, y2, color='r', linewidth = 3, linestyle = '-.')
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, color='g', linewidth = 3, linestyle = ':')
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")

plt.subplot(1,2,2)
plt.plot(x, y2, color='r', linewidth = 3, linestyle = '-.')
plt.title("Linear relation graph")
plt.xlabel ("X")
plt.ylabel ("Y")
plt.grid()
plt.show()
```

```python
details = {'Madhu': 96, 'Sanika': 90, 'Kajal': 65}
```

```python
Names = list(details.keys())
```

```python
Marks = list(details.values())
```

```python
plt.bar(Names, Marks)
plt.show
```

```python
plt.bar(Names, Marks)
plt.title("Marks distribution")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.show
```

```python
plt.barh(Names, Marks, color = 'r')
plt.title("Distribution of marks of students")
plt.xlabel("Marks")
plt.ylabel("Names")
plt.grid()
plt.show()
```
